```python
import pandas as pd, numpy as np, json, os
```

```python
m=json.loads(open('json/member_timelines.json','r').read())
```

```python
m=m[:1000]
```

```python
data={'Ország':{},'Bizottság':{},'Párt':{}}
attention={}
for k,i in enumerate(m):
    if k%1000==0: print(k/len(m)*100,'%')
    if i['Típus'] in list(data.keys()):
        if i['Csoportok'] not in data[i['Típus']]:data[i['Típus']][i['Csoportok']]={}
        for t in pd.date_range(i['start'],i['end']):
            st=str(t)[:10]
            if st not in data[i['Típus']][i['Csoportok']]:data[i['Típus']][i['Csoportok']][st]=set()
            data[i['Típus']][i['Csoportok']][st].add(i['Name'])
            
            if i['Name'] not in attention:attention[i['Name']]={}
            if st not in attention[i['Name']]:attention[i['Name']][st]=0
            attention[i['Name']][st]+=1
```

```python
edges={}
points={'Ország':2,'Bizottság':3,'Párt':1}
for tipus in data:
    for csoport in data[tipus]:
        print(csoport)
        for t in data[tipus][csoport]:
            names=list(data[tipus][csoport][t])
            anneal=max(0.1,-np.log(float((pd.to_datetime('2019-05-01')-pd.to_datetime(t))/(pd.to_datetime('2019-05-01')-pd.to_datetime('1989-01-01')))))
            for r in range(len(names)):
                name=names[r]
                for p in range(r+1,len(names)):
                    name2=names[p]
                    v=points[tipus]
                    nset=tuple(np.sort([name,name2]))
                    if nset not in edges:edges[nset]={'anneal':0,'values':0}
                    edges[nset]['values']+=v
                    edges[nset]['anneal']+=anneal
```

Network analysis

```python
import networkx as nx
```

```python
G=nx.Graph(list(edges.keys()))
```

nodes

```python
pr = nx.pagerank(G)
nodes={i:{'id':i,'pr':pr[i]/np.mean(list(pr.values()))} for i in pr}
```

```python
from networkx.algorithms import community
communities_generator = community.girvan_newman(G)
for i in range(8):
    communities = next(communities_generator)
for i,names in enumerate(sorted(map(sorted, communities))):
    for name in names:
        nodes[name]['community']='c'+str(i)
```

```python
from networkx.algorithms.community import k_clique_communities
communities = list(k_clique_communities(G, 7))
for i,names in enumerate(sorted(map(sorted, communities))):
    for name in names:
        nodes[name]['kclique']='c'+str(i)
```

```python
from networkx.algorithms.community import greedy_modularity_communities
communities = list(greedy_modularity_communities(G))
for i,names in enumerate(sorted(map(sorted, communities))):
    for name in names:
        nodes[name]['modularity']='c'+str(i)
```

```python
bc=nx.betweenness_centrality(G) #weight
for name in bc:
    nodes[name]['bc']=bc[name]/np.mean(list(bc.values()))
```

```python
dc=nx.degree_centrality(G)
for name in dc:
    nodes[name]['dc']=dc[name]/np.mean(list(dc.values()))
```

```python
ec=nx.eigenvector_centrality(G) #weight
for name in ec:
    nodes[name]['ec']=ec[name]/np.mean(list(ec.values()))
```

links

```python
ebc=nx.edge_betweenness_centrality(G) #weight
```

```python
links=[]
for i,nset in enumerate(edges):
    w=edges[nset]['values']/np.nanmean([i['values'] for i in edges.values()])
    nset=list(nset)
    if w>0:
        if tuple(nset) in ebc:
            u=ebc[tuple(nset)]/np.mean(list(ebc.values()))
        elif tuple(nset)[::-1] in ebc: 
            u=ebc[tuple(nset)[::-1]]/np.mean(list(ebc.values()))
        else:
            u=1
        links.append({'source':nset[0],'target':nset[1],
                      'value':w,
                      'ebc':u
                     })
print(len(links))
```

save

```python
open('json/network_pagerank_normed.json','w').write(json.dumps({'nodes':nodes,'links':links}))
```
