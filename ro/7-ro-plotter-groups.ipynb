```python
import pandas as pd, numpy as np, json, os
import matplotlib.pyplot as plt
```

```python
base_url='http://www.cdep.ro'
```

```python
efs4=json.loads(open('json/member_timelines.json','r').read())
```

```python
efs4=pd.DataFrame(efs4)
```

```python
top15=efs4[efs4['Típus']=='Ország'].groupby(['Csoportok']).count()[['Name']].sort_values('Name',ascending=False)[:15].index
```

```python
df=efs4[efs4['Típus']=='Ország'].groupby(['Csoportok','start']).count()[['Name']].loc[top15].reset_index(0)
```

```python
df.index=pd.to_datetime(df.index)
```

```python
df=df.sort_index()
df=df.set_index('Csoportok',append=True).unstack()
```

```python
df.plot(figsize=(14,8))
```
