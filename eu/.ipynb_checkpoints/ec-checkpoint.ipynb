```python
import rdflib
import gzip

g = rdflib.Graph()
result = g.parse(gzip.open("ec/VotingResults.ttl.gz"), format='turtle')
#http://data.consilium.europa.eu/data/public-voting/council-votes-on-legislative-acts.zip

print("graph has %s statements." % len(g))
# prints graph has 79 statements.

for subj, pred, obj in g:
   if (subj, pred, obj) not in g:
       raise Exception("It better be!")

s = g.serialize(format='n3')
```
